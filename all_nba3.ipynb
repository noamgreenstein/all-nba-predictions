```python
from nba_api.stats.endpoints import leaguedashplayerstats
import pandas as pd
player_stats = leaguedashplayerstats.LeagueDashPlayerStats(season='2020-21', measure_type_detailed_defense='Advanced').get_data_frames()[0]
pd.set_option('display.max_columns', None)
player_stats.head()
```

```python
player_stats.columns
cols_using = ['PLAYER_ID', 'PLAYER_NAME', 'TEAM_ID', 'TEAM_ABBREVIATION',
       'AGE', 'GP_RANK','W_RANK', 'L_RANK', 'W_PCT_RANK', 'MIN_RANK', 'E_OFF_RATING_RANK',
       'OFF_RATING_RANK', 'sp_work_OFF_RATING_RANK', 'E_DEF_RATING_RANK',
       'DEF_RATING_RANK', 'sp_work_DEF_RATING_RANK', 'E_NET_RATING_RANK',
       'NET_RATING_RANK', 'sp_work_NET_RATING_RANK', 'AST_PCT_RANK',
       'AST_TO_RANK', 'AST_RATIO_RANK', 'OREB_PCT_RANK', 'DREB_PCT_RANK',
       'REB_PCT_RANK', 'TM_TOV_PCT_RANK', 'E_TOV_PCT_RANK', 'EFG_PCT_RANK',
       'TS_PCT_RANK', 'USG_PCT_RANK', 'E_USG_PCT_RANK', 'E_PACE_RANK',
       'PACE_RANK', 'sp_work_PACE_RANK', 'PIE_RANK', 'FGM_RANK', 'FGA_RANK',
       'FGM_PG_RANK', 'FGA_PG_RANK', 'FG_PCT_RANK']
```

```python
import os
import glob
player_df = pd.DataFrame()

main_folder_path = 'Teams'
for subfolder in os.listdir(main_folder_path):
    subfolder_path = os.path.join(main_folder_path, subfolder)
    if os.path.isdir(subfolder_path):
        print(f"Processing files in: {subfolder}")
        file_pattern = os.path.join(subfolder_path, '*.csv')
        for file_path in glob.glob(file_pattern):
            player_df = pd.concat([player_df, pd.read_csv(file_path)])
                
```

```python
player_df.head()
```

```python
player_df = player_df.drop(['Unnamed: 0'], axis=1)
player_df['TEAM_ID'] = player_df['TEAM_ID'].apply(lambda x: str(x)[:-2])
player_df.head()
```

```python
player_df['PLAYER_ID'] = player_df['PLAYER_ID'].apply(lambda x: str(x)[:-2])
player_df.head()
```

```python
player_stats = pd.DataFrame()
for year in range(2000, 2023):
    season = f'{year}-{str(year+1)[-2:]}'
    temp = leaguedashplayerstats.LeagueDashPlayerStats(season=season, measure_type_detailed_defense='Advanced').get_data_frames()[0]
    temp = temp[cols_using]
    temp['SEASON_ID'] = [season for _ in range(temp.shape[0])]
    player_stats = pd.concat([player_stats, temp])
    print(year)
    
```

```python
player_stats['PLAYER_ID'] = player_stats['PLAYER_ID'].astype(str)
```

```python
player_df = pd.merge(left=player_df, right=player_stats, on=['PLAYER_ID', 'SEASON_ID'], how='left')
```
