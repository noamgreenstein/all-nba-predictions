```python
from nba_api.stats.static import teams
teams = teams.get_teams()
```

```python
from nba_api.stats.endpoints import commonteamroster

rosters = {}
for team in teams:
    rosters[team['abbreviation']] = []
    print(team['full_name'])
    for year in range(2000, 2023):
         rosters.get(team['abbreviation']).append({year: commonteamroster.CommonTeamRoster(team_id=team['id'], season=str(year)).get_data_frames()[0]})
         print(year)
```

```python
player_avgs = {}
```

```python
def get_avgs(team):
    player_stats = {}
    player_avgs[team] = {}
    years = rosters.get(team)
    for year in years:
        year_str = list(year.keys())[0]
        season = f'{year_str}-{str(year_str+1)[-2:]}'
        player_avgs.get(team)[year_str] = []
        players = list(year.values())[0]
        for index in players.index:
            player_id = players.at[index, 'PLAYER_ID']
            if player_id in player_stats.keys():
                career = player_stats.get(player_id)
            else:
                career = playercareerstats.PlayerCareerStats(player_id=player_id).get_data_frames()[0]
                player_stats[player_id] = career
            player_avgs.get(team).get(year_str).append(career[career['SEASON_ID'] == season])
        print(season)
    print(team)
```

```python
import pandas as pd
for team in player_avgs.keys():
    for year in player_avgs[team].keys():
        player_dfs = player_avgs[team][year]
        df = pd.DataFrame()
        for player in player_dfs:
            df = pd.concat([df, player])
        df.to_csv(f'Teams/{team}_{year}.csv')
```

```python
player_avgs = {}

def save_stats(team):
    for year in player_avgs[team].keys():
        player_dfs = player_avgs[team][year]
        df = pd.DataFrame()
        for player in player_dfs:
            df = pd.concat([df, player])
        df.to_csv(f'Teams/{team}/{team}_{year}.csv')
    print(f'{team} Saved')


for team in list(rosters.keys())[5:10]:
    get_avgs(team)
    save_stats(team)
```

```python
import os
import shutil

# Path to the directory with files
directory = 'Teams'

# Loop through all files in the directory
for filename in os.listdir(directory):
    if os.path.isfile(os.path.join(directory, filename)):
        # Get the first letter of the file name
        team = filename[:3]

        # Create a new directory path
        new_directory = os.path.join(directory, team)

        # Create a new directory if it doesn't exist
        if not os.path.exists(new_directory):
            os.makedirs(new_directory)

        # Move the file to the new directory
        shutil.move(os.path.join(directory, filename), os.path.join(new_directory, filename))
```
